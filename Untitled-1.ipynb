```python
import torch
import numpy as np
import pandas as pd
import os

from utils import  fix_randomness
import argparse

from trainers.abstract_trainer import  AbstractTrainer
```

```python
parser = argparse.ArgumentParser()

parser.add_argument('--save_dir',   default='experiments_logs',         type=str, help='Directory containing all experiments')
parser.add_argument('--da_method',  default='CoDATS',                   type=str, help='NO_ADAPT, DANN, CoDATS, CoTMix, TARGET_ONLY')
parser.add_argument('--data_path',  default=r'/home/jabdullayev/phd/datasets/RehabPile/classification/',               type=str, help='Path containing datase2t')
parser.add_argument('--dataset',    default='KERAAL_clf_mc_ELK',         type=str, help='Dataset of choice: (WISDM - EEG - HAR - HHAR_SA)')
parser.add_argument('--backbone',   default='LITEMV',                   type=str, help='Backbone of choice: (LITEMV - CNN - RESNET18 - TCN)')
parser.add_argument('--num_runs',   default=5,                          type=int, help='Number of consecutive run with different seeds')
parser.add_argument('--device',     default= "cuda",                    type=str, help='cpu or gpu')


# arguments
args = parser.parse_args([])
    
trainer = AbstractTrainer(args)

for fold_id in range(0, 1):
    for run_id in range(1, 5):
        # Fixin randome seeds
        fix_randomness(run_id)
        
        # Load data
        trainer.load_data(fold_number=fold_id, test_subject_id=None)
        trainer.initialize_algorithm()                
        
        pretrained_model_log_dir = os.path.join('experiments_logs/', args.dataset, "CoDATS", "fold_" + str(fold_id), "run_" + str(run_id))
        first_chk, last_chk, best_chk = trainer.load_checkpoint(pretrained_model_log_dir)

        trainer.algorithm.network.load_state_dict(last_chk)
        
        
        src_feats_all, src_labels_all, src_subjects_all = [], [], []
        for src_x, src_y, src_s in trainer.train_dl:            
            src_x, src_y, src_s = src_x.to('cuda'), src_y.to('cuda'), src_s.to('cuda')              # extract source features

            src_feat = trainer.algorithm.feature_extractor(src_x)
            
            src_feats_all.append(src_feat)
            src_labels_all.append(src_y)
            src_subjects_all.append(src_s)
            
            # print(f'src feature shape: {src_feat.shape}')

        src_feats_all = torch.cat(src_feats_all, dim=0)
        src_labels_all = torch.cat(src_labels_all, dim=0)
        src_subjects_all = torch.cat(src_subjects_all, dim=0)
        
        
        print(f'src_feats_all shape: {src_feats_all.shape}')
        
        
                
        
        trg_feats_all, trg_labels_all, trg_subjects_all = [], [], []
        for trg_x, trg_y, trg_s in trainer.test_dl:            
            trg_x, trg_y, trg_s = trg_x.to('cuda'), trg_y.to('cuda'), trg_s.to('cuda')              # extract source features

            trg_feat = trainer.algorithm.feature_extractor(trg_x)
            trg_feats_all.append(trg_feat)
            trg_labels_all.append(trg_y)
            trg_subjects_all.append(trg_s)
            
            
            # print(f'src feature shape: {src_feat.shape}')

        trg_feats_all = torch.cat(trg_feats_all, dim=0)
        trg_labels_all = torch.cat(trg_labels_all, dim=0)
        trg_subjects_all = torch.cat(trg_subjects_all, dim=0)
        
        
        print(f'trg_feats_all shape: {trg_feats_all.shape}')
        
        
        
        break
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

def visualize_with_class_and_subjects(
    source_features, target_features,
    source_labels, target_labels,
    source_subjects, target_subjects,
    num_classes=4,
    annotate=False,           # draw subject IDs as text near points (static)
    max_annot=200,            # cap text annotations to avoid clutter
    random_state=42
):
    # combine
    Xs, Xt = np.asarray(source_features), np.asarray(target_features)
    ys, yt = np.asarray(source_labels),   np.asarray(target_labels)
    ss, st = np.asarray(source_subjects), np.asarray(target_subjects)

    X   = np.vstack([Xs, Xt])
    y   = np.concatenate([ys, yt])
    dom = np.array([0]*len(Xs) + [1]*len(Xt))
    subj= np.concatenate([ss, st]).astype(str)

    # safe perplexity for tiny target
    n = len(X)
    perplexity = max(2, min(30, n - 1))
    Z = TSNE(n_components=2, perplexity=perplexity, init="pca",
             learning_rate="auto", random_state=random_state).fit_transform(X)

    colors = plt.cm.tab10(np.linspace(0, 1, num_classes))
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111)

    # draw class+domain
    handles = []
    for c in range(num_classes):
        m_src = (y == c) & (dom == 0)
        m_tgt = (y == c) & (dom == 1)

        h1 = ax.scatter(Z[m_src,0], Z[m_src,1], s=50, alpha=0.6,
                        c=[colors[c]], edgecolors='black', linewidth=0.5,
                        label=f"Source class {c}")
        h2 = ax.scatter(Z[m_tgt,0], Z[m_tgt,1], s=100, alpha=0.9,
                        c=[colors[c]], marker='s', edgecolors='black', linewidth=1,
                        label=f"Target class {c}")
        handles.extend([h1, h2])

    ax.set_xlabel('t-SNE Dim 1'); ax.set_ylabel('t-SNE Dim 2')
    ax.set_title('t-SNE: Classes, Domains, and Subjects')
    # de-duplicate legend labels
    by_label = {}
    for h in handles:
        by_label[h.get_label()] = h
    ax.legend(by_label.values(), by_label.keys(), fontsize=8, ncol=2)
    ax.grid(True, alpha=0.3)

    # optional text annotations (subject ids)
    if annotate:
        # avoid overcrowding: annotate up to max_annot (prioritize target first)
        idx_order = np.r_[np.where(dom==1)[0], np.where(dom==0)[0]]
        idx_order = idx_order[:max_annot]
        for i in idx_order:
            ax.text(Z[i,0], Z[i,1], subj[i],
                    fontsize=7, ha='left', va='bottom', alpha=0.85)

    plt.show()
```

```python
visualize_with_class_and_subjects(
    src_feats_all.detach().cpu(), trg_feats_all.detach().cpu(),
    src_labels_all.detach().cpu(), trg_labels_all.detach().cpu(),
    src_subjects_all.detach().cpu(), trg_subjects_all.detach().cpu(),
    num_classes=4,
    annotate=True,      # set False if too cluttered
    max_annot=120
)
```
